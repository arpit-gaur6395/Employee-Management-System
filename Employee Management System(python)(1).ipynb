```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import mysql.connector as mysql

class EmployeeManagementSystem:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1550x1250')
        self.root.title('Employee Management System')
        self.login_page()

    def login_page(self):
        self.login_frame = Frame(self.root, bd=2, relief=RIDGE, bg='orange')
        self.login_frame.place(x=550, y=250, width=450, height=300)

        title = Label(self.login_frame, text='Login', font=('times new roman', 20, 'bold'), fg='black', bg='yellow')
        title.pack(side=TOP, fill=X)

        username_label = Label(self.login_frame, text='Username', font=('times new roman', 15, 'bold'), bg='Light Blue')
        username_label.place(x=50, y=70)
        self.username_entry = ttk.Entry(self.login_frame, width=25, font=("Arial", 11, "bold"))
        self.username_entry.place(x=180, y=70)

        password_label = Label(self.login_frame, text='Password', font=('times new roman', 15, 'bold'), bg='Light Blue')
        password_label.place(x=50, y=130)
        self.password_entry = ttk.Entry(self.login_frame, width=25, font=("Arial", 11, "bold"), show='*')
        self.password_entry.place(x=180, y=130)

        login_button = Button(self.login_frame, text='Login', command=self.verify_login, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        login_button.place(x=150, y=200)

    def verify_login(self):
        username = self.username_entry.get()
        password = self.password_entry.get()
        if username == 'Arpit' and password == '0101':  # Simple credential check, replace with DB check in real applications
            self.login_frame.destroy()
            self.main_page()
        else:
            messagebox.showerror('Error', 'Invalid Credentials')

    def main_page(self):
        try:
            self.con = mysql.connect(host="localhost", user="root", password="21112002", database="employee_management")
            self.c = self.con.cursor()
            self.create_table()
        except mysql.Error as e:
            print(f"Error connecting to MySQL: {e}")

        title = Label(self.root, text='Employee Management System', font=('times new roman', 25, 'bold'), fg='black', bg='yellow')
        title.place(x=0, y=0, width=1530, height=50)

        main_frame = Frame(self.root, bd=2, relief=RIDGE)
        main_frame.place(x=0, y=50, width=1530, height=670)

        Uframe = LabelFrame(main_frame, bd=2, relief=RIDGE, bg='Light Blue', text='Employee Information')
        Uframe.place(x=0, y=0, width=1530, height=450)



        lb = Label(Uframe, text='Employee Name', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=0, column=0, padx=5, pady=7, sticky=W)
        self.employee_name = ttk.Entry(Uframe, width=25, font=("Arial", 11, "bold"))
        self.employee_name.grid(row=0, column=1, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Mobile No', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=1, column=0, padx=5, pady=7, sticky=W)
        self.mobile_no = ttk.Entry(Uframe, width=25, font=("Arial", 11, "bold"))
        self.mobile_no.grid(row=1, column=1, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='E-mail', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=2, column=0, padx=5, pady=7, sticky=W)
        self.e_mail = ttk.Entry(Uframe, width=25, font=("Arial", 11, "bold"))
        self.e_mail.grid(row=2, column=1, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Gender', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=3, column=0, padx=5, pady=7, sticky=W)
        self.gender = ttk.Combobox(Uframe, width=30, state='readonly')
        self.gender['values'] = ('Select', 'Male', 'Female', 'Other')
        self.gender.current(0)
        self.gender.grid(row=3, column=1, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Department', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=4, column=0, padx=5, pady=7, sticky=W)
        self.department = ttk.Combobox(Uframe, width=30, state='readonly')
        self.department['values'] = ('Select', 'Engineering', 'Management', 'Testing', 'Maintenance')
        self.department.current(0)
        self.department.grid(row=4, column=1, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Designation', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=0, column=4, padx=5, pady=7, sticky=W)
        self.designation = ttk.Combobox(Uframe, width=30, state='readonly')
        self.designation['values'] = ('Select', 'Software Engineer', 'Data Engineer', 'Design Engineer', 'Testing Engineer', 'Cyber Security Engineer')
        self.designation.current(0)
        self.designation.grid(row=0, column=5, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Address', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=1, column=4, padx=4, pady=7, sticky=W)
        self.address = ttk.Entry(Uframe, width=25, font=("Arial", 11, "bold"))
        self.address.grid(row=1, column=5, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Experience', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=2, column=4, padx=5, pady=7, sticky=W)
        self.experience = ttk.Combobox(Uframe, width=30, state='readonly')
        self.experience['values'] = ('Select', 'Yes', 'No')
        self.experience.current(0)
        self.experience.grid(row=2, column=5, sticky=W, padx=5, pady=7)

        lb = Label(Uframe, text='Salary', font=('times new roman', 18, 'bold'), bg='Light Blue')
        lb.grid(row=3, column=4, padx=5, pady=7, sticky=W)
        self.salary = ttk.Entry(Uframe, width=25, font=("Arial", 11, "bold"))
        self.salary.grid(row=3, column=5, sticky=W, padx=5, pady=7)

        btn_frame = Frame(main_frame, bd=2, relief=RIDGE, bg='white')
        btn_frame.place(x=0, y=450, width=1530, height=150)

        add_btn = Button(btn_frame, text='Add', command=self.add_employee_to_db, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        add_btn.grid(row=0, column=0, padx=5, pady=10)

        clear_btn = Button(btn_frame, text='Clear', command=self.clear_fields, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        clear_btn.grid(row=0, column=1, padx=5, pady=10)

        display_btn = Button(btn_frame, text='Display', command=self.display_data, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        display_btn.grid(row=0, column=2, padx=5, pady=10)

        delete_btn = Button(btn_frame, text='Delete', command=self.delete_employee, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        delete_btn.grid(row=0, column=3, padx=5, pady=10)

        view_btn = Button(btn_frame, text='View Information', command=self.view_employee, width=18, font=('times new roman', 12, 'bold'), bg='blue', fg='white')
        view_btn.grid(row=0, column=4, padx=5, pady=10)

        self.tree_frame = Frame(main_frame, bd=2, relief=RIDGE, bg='white')
        self.tree_frame.place(x=0, y=550, width=1530, height=120)

        self.tree = ttk.Treeview(self.tree_frame, columns=("name", "mobile", "email", "gender", "department", "designation", "address", "experience", "salary"), show='headings')
        self.tree.heading("name", text="Employee Name")
        self.tree.heading("mobile", text="Mobile No")
        self.tree.heading("email", text="E-mail")
        self.tree.heading("gender", text="Gender")
        self.tree.heading("department", text="Department")
        self.tree.heading("designation", text="Designation")
        self.tree.heading("address", text="Address")
        self.tree.heading("experience", text="Experience")
        self.tree.heading("salary", text="Salary")
        self.tree.pack(fill=BOTH, expand=1)

    def create_table(self):
        try:
            self.c.execute('''
                CREATE TABLE IF NOT EXISTS ems(
                    employee_name VARCHAR(100),
                    mobile_no VARCHAR(100) PRIMARY KEY,
                    e_mail VARCHAR(100),
                    gender VARCHAR(100),
                    department VARCHAR(100),
                    designation VARCHAR(100),
                    address VARCHAR(100),
                    experience VARCHAR(100),
                    salary VARCHAR(100)
                )
            ''')
            self.con.commit()
        except mysql.Error as e:
            print(f"Error creating table: {e}")

    def add_employee_to_db(self):
        try:
            employee_name = self.employee_name.get()
            mobile_no = self.mobile_no.get()
            e_mail = self.e_mail.get()
            gender = self.gender.get()
            department = self.department.get()
            designation = self.designation.get()
            address = self.address.get()
            experience = self.experience.get()
            salary = self.salary.get()
            
            if employee_name and mobile_no and e_mail and gender != 'Select' and department != 'Select' and designation != 'Select' and experience != 'Select':
                self.c.execute("INSERT INTO ems (employee_name, mobile_no, e_mail, gender, department, designation, address, experience, salary) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)",
                               (employee_name, mobile_no, e_mail, gender, department, designation, address, experience, salary))
                self.con.commit()
                print("Employee added successfully!")
                self.clear_fields()
            else:
                print("Please fill all fields!")
        except mysql.Error as e:
            print(f"Error adding employee: {e}")

    def clear_fields(self):
        self.employee_name.delete(0, END)
        self.mobile_no.delete(0, END)
        self.e_mail.delete(0, END)
        self.gender.current(0)
        self.department.current(0)
        self.designation.current(0)
        self.address.delete(0, END)
        self.experience.current(0)
        self.salary.delete(0, END)
    
    def display_data(self):
        try:
            self.c.execute("SELECT * FROM ems")
            rows = self.c.fetchall()
            if rows:
                self.tree.delete(*self.tree.get_children())
                for row in rows:
                    self.tree.insert("", END, values=row)
        except mysql.Error as e:
            print(f"Error fetching data: {e}")

    def delete_employee(self):
        selected_item = self.tree.selection()
        if selected_item:
            employee_data = self.tree.item(selected_item, 'values')
            try:
                self.c.execute("DELETE FROM ems WHERE mobile_no = %s", (employee_data[1],))
                self.con.commit()
                print("Employee deleted successfully!")
                self.display_data()
            except mysql.Error as e:
                print(f"Error deleting employee: {e}")
        else:
            print("Please select an employee to delete.")

    def view_employee(self):
        selected_item = self.tree.selection()
        if selected_item:
            employee_data = self.tree.item(selected_item, 'values')
            self.clear_fields()
            self.employee_name.insert(0, employee_data[0])
            self.mobile_no.insert(0, employee_data[1])
            self.e_mail.insert(0, employee_data[2])
            self.gender.set(employee_data[3])
            self.department.set(employee_data[4])
            self.designation.set(employee_data[5])
            self.address.insert(0, employee_data[6])
            self.experience.set(employee_data[7])
            self.salary.insert(0, employee_data[8])
        else:
            print("Please select an employee to view.")

if __name__ == "__main__":
    root = Tk()
    ems = EmployeeManagementSystem(root)
    root.mainloop()
```
